```python
from gensim.models import Word2Vec
from nltk.cluster import KMeansClusterer
import nltk
from sklearn import cluster
from sklearn import metrics
import numpy as np
from nltk.tokenize import word_tokenize
 
# training data from the corpus of skills extracted from our resume dataset
 
g = open("ANTdata.txt", "r")
sentences = []
for x in g:
    sentences.append(word_tokenize(x))
g.close()    
```

```python
# training model
#Learning a word embedding from text involves loading and 
#organizing the text into sentences and providing them to
#the constructor of a new Word2Vec() instance.
model = Word2Vec(sentences, min_count=1, size = 300)
X = model[model.wv.vocab]
```

```python
#skills clustering algorithm
NUM_CLUSTERS=10
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25)
kmeans = cluster.KMeans(n_clusters=NUM_CLUSTERS)
kmeans.fit(X)
 
#parameters to gauge clustering algo 
#print ("Cluster id labels for inputted data")
#print (kmeans.labels_)
#print ("Centroids data")
#print (kmeans.cluster_centers_)
 
#to measure how similar an object is to its own cluster (cohesion) compared to other clusters (separation)
#silhouette_score = metrics.silhouette_score(X, labels, metric='euclidean') 
#print ("Silhouette_score: ")
#print (silhouette_score)
```

```python
# to check the clusters assigned to each skill
# and to produce a dictionary wherein keys are the words (skills eg python, java) and values are the vector arrays eg [0,0,0,0,1,0,0,0,1,0]
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
skills_dict = {}
y = np.zeros(NUM_CLUSTERS)
words = list(model.wv.vocab)
for i, word in enumerate(words):  
    print (word + ":" + str(assigned_clusters[i]))
    y[assigned_clusters[i]] = 1
    skills_dict[word]=list(y)
    y = np.zeros(NUM_CLUSTERS)  
```
